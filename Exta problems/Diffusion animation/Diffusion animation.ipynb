```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook
from random import randrange
fig = plt.figure()
plt.axis([-100,100, -100,100])
m=10
r=[]
for i in range(-m,m+1):
    for j in range(-m,m+1):
        r.append([i,j])
L=100
m = 10
N = ((2*m)+1)**2
global r,ra,N,m,L
ra = np.array(r)
particle, = plt.plot(ra[:,0],ra[:,1],'o', color='green')

def update(i):
    
    for k in range(N):
        n = randrange(N)

        x0 = r[n][0]
        y0 = r[n][1]
        p = randrange(4)
        if p == 0 :
            r[n][0] = x0+1
        elif p == 1 :
            r[n][1] = y0+1

        elif p == 2 :
            r[n][0] = x0-1

        else :
            r[n][1] = y0-1

        if  r[n][0] == L or r[n][1] == L or r[n][0] == -L or r[n][1] == -L :

            r[n][0] = x0
            r[n][1] = y0
            
    ra = np.array(r)
    particle.set_data(ra[:,0],ra[:,1])
    return particle,
plt.xlabel("X")
plt.ylabel("Y")
ani = FuncAnimation(fig, update, frames=range(500), interval=1)
```
