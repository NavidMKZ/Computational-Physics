```python
from tkinter import *
cycles = [["e"],[1,2],[2,3],[1,3,2],[1,2,3],[1,3],[3,4],
          [[1,2],[3,4]],[2,4,3],[1,4,3,2],[1,2,4,3],[1,4,3],[2,3,4],
          [1,3,4,2],[2,4],[1,4,2],[[1,3],[2,4]],[1,4,2,3],[1,2,3,4],
          [1,3,4],[1,2,4],[1,4],[1,3,2,4],[[1,4],[2,3]]]
A4 = [1,2,3,4]

def permutation(cycle,A4):
    A = []
    A = A4.copy()
    if cycle[0] == "e":
        return ["%s"%(cycle[0]),A4,A]
    
    elif len(cycle) == 3 :
        index2 = A.index(cycle[1])
        index3 = A.index(cycle[2])
        A[index2] , A[index3] =  A[index3] , A[index2]
        index1 = A.index(cycle[0])
        index2 = A.index(cycle[1])
        A[index1] , A[index2] =  A[index2] , A[index1]
        s = "P" + "%i%i%i"%(cycle[0],cycle[1],cycle[2])
        return [s,A4,A]
    
    elif len(cycle) == 4:
        index3 = A.index(cycle[2])
        index4 = A.index(cycle[3])
        A[index4] , A[index3] =  A[index3] , A[index4]
        index2 = A.index(cycle[1])
        index3 = A.index(cycle[2])
        A[index3] , A[index2] =  A[index2] , A[index3]
        index1 = A.index(cycle[0])
        index2 = A.index(cycle[1])
        A[index2] , A[index1] =  A[index1] , A[index2]
        
        s = "P" + "%i%i%i%i"%(cycle[0],cycle[1],cycle[2],cycle[3])
        return [s,A4,A]
        
    try:
        index1 = A.index(cycle[1][0])
        index2 = A.index(cycle[1][1])
        A[index1] , A[index2] =  A[index2] , A[index1]
        index1 = A.index(cycle[0][0])
        index2 = A.index(cycle[0][1])
        A[index1] , A[index2] =  A[index2] , A[index1]
        s = "P" + "(%i%i)"%(cycle[0][0],cycle[0][1]) + "(%i%i)"%(cycle[1][0],cycle[1][1])
        return [s,A4,A]
    
    except TypeError:
        index1 = A.index(cycle[0])
        index2 = A.index(cycle[1])
        A[index1] , A[index2] =  A[index2] , A[index1]
        s = "P" + "%i%i"%(cycle[0],cycle[1])
        return [s,A4,A]
    
S4 = []
for cycle in cycles:
    S4.append(permutation(cycle,A4))

head = ["S4"]
for name in S4:
    head.append(name[0])

def tracing(S4,p2):
    for p in S4:
        if p2[2] == p[2]:
            return p[0]

P = []     
for index,pi in enumerate(cycles):
    pij = [S4[index][0]]
    for pj in cycles:
        
        p1 , p2 = list(),list()
        p1 = permutation(pj,A4)
        p2 = permutation(pi,p1[2])
        
        pij.append(tracing(S4,p2))
    P.append(pij)
    
lst = [(head)]
for i in P:
    lst.append((i))
    
class Table:
    def __init__(self,root):

        for i in range(total_rows):
            for j in range(total_columns):
                self.e = Entry(root, width=8, fg='black',font=('Arial',10))
                self.e.grid(row=i, column=j)
                self.e.insert(END, lst[i][j])


total_rows = len(lst)
total_columns = len(lst[0])

root = Tk()
t = Table(root)
root.mainloop()

lst = []
for pi in S4:
    lst.append((pi[0],pi[2]))
total_rows = len(lst)
total_columns = len(lst[0])

root = Tk()
t = Table(root)
root.mainloop()
```
