```python
import numpy as np
import matplotlib.pyplot as plt
def d_dx2(phi, x):
    ret = -2*phi
    ret[:-1] += phi[1:]
    ret[1:]+= phi[:-1]
    return ret

def d_dt(phi, x, D):
    return D*d_dx2(phi,x)

D  = 0.05
n_stories = 500
t_max = 100
bound = 7
mesh = 2*bound
x = np.linspace(-bound, bound, mesh+1)
# trange = np.arange(t_max)
phi_t0 = np.zeros((mesh+1))
phi_t0[int(mesh/2)] = n_stories
phi = np.zeros((t_max, mesh+1))
phi[0, :] = phi_t0

for t in range(t_max-1):
    phi[t+1, :] = phi[t,:] + d_dt(phi[t,:],x, D)
    
import matplotlib.pyplot as plt
colorinterpolation = 50
colourMap = plt.cm.jet
# Set meshgrid
X, Y = np.meshgrid(np.arange(-7, 8), np.arange(0, t_max))
plt.contourf(X, Y, phi, colorinterpolation, cmap=colourMap)
plt.xlabel("X")
plt.ylabel("Time")
# Set Colorbar
plt.colorbar()
# Show the result in the plot window
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
Q = 0.2
pi = np.pi
ep = 8.85*(10**(-12))
r_ring = 0.8

angle_mesh = 361
radius_mesh = 101

vGrid = np.zeros((radius_mesh, angle_mesh))

r_max = 100
a_max = 360
r = np.linspace(0, r_max, radius_mesh)
t = np.linspace(0, a_max, angle_mesh)

dr = r[1] - r[0]
dt = t[1] - t[0]
 
maxItr = 1500
Itr = 0
tolerance = 0.001
diff=np.empty((angle_mesh,angle_mesh))

while Itr < maxItr and len(diff)>0:
    vTemp = vGrid.copy()
    for i in range(1,radius_mesh-1):
        for j in range(0,angle_mesh-1):
            if r[i] == 80:
                term4 = Q / (2*pi*ep*r_ring)
                vGrid[i,j] = term4
            else:
                C = (1 / ((1/r[i]*dr) + (2/dr**2) + (2/(r[i]*dt)**2)))
                term1 = (1/r[i])*(vGrid[i+1,j]/dr)
                term2 = (1/dr**2)*(vGrid[i+1,j]+vGrid[i-1,j])
                term3 = (1/(r[i]*dt)**2)*(vGrid[i,j+1]+vGrid[i,j-1])
                vGrid[i,j] = C*(term1 + term2 + term3)
    Itr+=1
    diff = np.where(abs(vTemp-vGrid)>tolerance)
    


colorinterpolation = 50
colourMap = plt.cm.jet

radius_range = np.radians(np.linspace(0, a_max, angle_mesh))
angle_range = np.linspace(0, r_max, radius_mesh)
radius, angle = np.meshgrid(radius_range,angle_range)


fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
ax.set(title = "The potential of a ring of charge")
pt = ax.contourf(radius, angle, vGrid, colorinterpolation, cmap=colourMap)
fig.colorbar(pt)
plt.show()
```
